```python
import pandas as pd
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'
dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
dados_resultados = pd.read_csv('https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true')
#Esta função é utilizada para ler apenas as 5 primeiras linhas da nossa tabela
dados_resultados.head() 
```

```python
#Utilizando .unique conseguimos ver os valores unicos (variáveis) dentro da coluna
dados_resultados['acetylcholine_receptor_agonist'].unique()
```

```python
#A função select_dtypes nos permite selecionar a coluna por tipo (int64 = número inteiro), a função sum faz a soma
contagem_moa = dados_resultados.select_dtypes('int64').sum() 
contagem_moa
```

```python
#A função select_dtypes nos permite selecionar a coluna por tipo (int64 = número inteiro), a função sum faz a soma
#A função .sort_values serve para ordenarmos as resultados
contagem_moa = dados_resultados.select_dtypes('int64').sum() 
contagem_moa
```

```python
#Essa função serve para verificarmos as tipo de dados da nossa tabela
dados_resultados.info()
```

```python
#Utilizamos a função .drop para excluir uma coluna, neste caso a coluna id por ser um objeto
#Informamos o axis para que seja excluido a coluna toda e não apenas linhas
dados_resultados.drop('id', axis=1).sum().sort_values(ascending = False)
```

```python
#Somando linhas, a string pode ser somada, ela fica concatenada
dados_resultados.sum()
```

```python
#Somando uma string, verificando qntd de ações
dados_resultados.drop('id', axis=1).sum(axis=1) 
```

```python
#Esta função é utilizada para ler apenas as 5 primeiras linhas da nossa tabela
dados_resultados.head()
```

```python
#Utilizando a função axis vamos criar uma nova coluna n tabela dados_resultados com a soma dos mecanismos de ações ativos (n_moa)
dados_resultados['n_moa'] = dados_resultados.drop('id', axis=1).sum(axis=1) 
```

```python
#Validando as informações da coluna que criamos, assim podemos saber quais ativos foram ou não ativados
dados_resultados['n_moa'] != 0 
```

```python
#Criamos mais uma coluna para fazermos a validação dos dados, se o n_moa for diferente de 0 é True (ativado) e se for False (não ativado)
dados_resultados['ativo_moa'] = (dados_resultados['n_moa'] != 0) 
#Esta função é utilizada para ler apenas as 5 primeiras linhas da nossa tabela
dados_resultados.head() 
```

```python
#A função merge possibilita criar colunas com as informações de duas tabelas (similar ao procv no excel)
dados_combinados = pd.merge(dados, dados_resultados[['id', 'n_moa', 'ativo_moa']], on = 'id')
#Esta função é utilizada para ler apenas as 5 primeiras linhas da nossa tabela
dados_combinados.head()
```

```python
#A função query possibilita filtarmos alguma informação na tabela, nesse caso estamos filtrando os casos com_controle da coluna tratamento
dados_combinados.query('tratamento == "com_controle"') 
```

```python
#A função query possibilita filtarmos alguma informação na tabela, nesse caso estamos filtrando os casos com_controle da coluna tratamento, para facilitar a visualização vamos utilizar o value_counts para realizar a contagem
dados_combinados.query('tratamento == "com_controle"') ['ativo_moa'].value_counts() 
```

```python
#A função query possibilita filtarmos alguma informação na tabela, nesse caso estamos filtrando os casos com_droga da coluna tratamento, para facilitar a visualização vamos utilizar o value_counts para realizar a contagem
dados_combinados.query('tratamento == "com_droga"') ['ativo_moa'].value_counts()
```

```python
#Importando matplotlib
import matplotlib.pyplot as plt
#Importando seaborn
import seaborn as sns
#Essa função serve para salvar as configurações de padrão de imagem do seaborn
sns.set()
```

```python
#Visualizando quais são os compostos principais (droga)
composto_principal = dados_combinados['droga'].value_counts().index[:5]
#Plotando Index
composto_principal
```

## **Desafios**

***Desafio 01: Encontrar o top 10 das ações do MOA (inibidor, agonista...).***

```python
#Separando a ação com a função i.split
sep = [i.split('_')[-1] for i in contagem_moa.index ]
#Plotando
sep
```

```python
#Criando um novo dataframe com as informações ajustadas de ações e quantidade
ndf_moa = pd.DataFrame(data={'Ação': sep ,'Contagem' : contagem_moa.values})
#Plotando
ndf_moa
```

```python
#Agrupando por quantidade de ações em ordem decrescente e selecionando os 10 primeiros
top10_qntd = ndf_moa.groupby(by=['Ação']).count()['Contagem'].sort_values(ascending=False)[:10]
#Ajustando tamanho do gráfico
plt.figure(figsize=(10,5))
#Definindo eixos e cores do gráfico
ax = sns.barplot(x = top10_qntd.values, y = top10_qntd.index, palette = 'brg')
#Defininfo título dos eixos
ax.set(xlabel = 'Quantidade', ylabel = 'Ações')
#Definindo título do gráfico
ax.set_title('Top 10 Ações em Quantidade', fontdict = {'fontsize':20})
#Plotando
plt.plot()
```

```python
#Agrupando por soma de ações em ordem decrescente e selecionando os 10 primeiros
top10_qntd = ndf_moa.groupby(by=['Ação']).sum()['Contagem'].sort_values(ascending=False)[:10]
#Ajustando tamanho do gráfico
plt.figure(figsize=(10,5))
#Definindo eixos e cores do gráfico
ax = sns.barplot(x = top10_qntd.values, y = top10_qntd.index, palette = 'gist_ncar')
#Defininfo título dos eixos
ax.set(xlabel = 'Quantidade', ylabel = 'Ações')
#Definindo título do gráfico
ax.set_title('Top 10 Ações em Utilizações', fontdict = {'fontsize':20})
#Plotando
plt.plot()
```

***Desafio 04: Estudar sobre combinações de DF https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html.***

Quando utilizamos o how = **'inner'** o merge traz apenas o que é comum nas duas tabelas, intersecção das chaves;

Quando utilizamos o how = **'left'** o merge traz apenas as chaves da primeira coluna;

Quando utilizamos o how = **'right'** o merge traz apenas as chaves da segunda tabela;

uando utilizamos o how = **'outer'** o merge traz as chaves de ambas as tabelas, união das chaves.

***Desafio 06: Descobrir se tem algum composto que dependendo da configuração do experimento, ativa ou não ativa algum MOA.***

```python
#Criando duas tabelas, uma com os MOAS ativos e uma com os inativos
ativa_moa = dados_combinados.query('ativo_moa == True')
nao_ativa_moa = dados_combinados.query('ativo_moa == False')
```

```python
#Utilizaremos a função merge para verificarmos se foi possível encontrar algum valor igual nas duas tabelas
ativa_nao_ativa = pd.merge(ativa_moa, nao_ativa_moa, on = ['droga'], how = 'outer', indicator = True)
```

```python
#Validando os dados com o value_counts, o qual retornou "both 0"
ativa_nao_ativa['_merge'].value_counts()
```
