## **Desafios**

***Desafio 01: Ordenar o gráfico countplot.***

```python
#Criando uma nova tabela com filtro das informações de cod_compostos dentro da coluna composto
new_table = dados.query('composto in @cod_compostos')
new_table
```

```python
#Imprimindo os dados de maneira decrescente
new_table['composto'].value_counts().sort_values(ascending = False)
```

```python
#Trazendo os dados de composto ordenado
ordenando = new_table['composto'].value_counts(ascending = False).index
ordenando
```

```python
#Utilizamos essa função para salvar o padrão de imagens do seaborn
sns.set()
#Ajustando tamanho do gráfico
plt.figure(figsize = (7,6))
#Plotando o gráfico ordenado por composto
ax = sns.countplot(x = 'composto', data =  dados.query('composto in @cod_compostos'), order = ordenando)
#Incluido título no gráfico
ax.set_title('Top 5 compostos')
plt.show()
```

***Desafio 02: Melhorar a visualização alterando o tamanho da fonte.***

```python
#Definindo eixos
x = 'composto'
y = dados.query('composto in @cod_compostos')
```

```python
#Utilizamos essa função para salvar o padrão de imagens do seaborn
sns.set()
#Definindo tamanho do gráfico
plt.figure(figsize = (7,5))
#Plotando o gráfico ordenado por composto e definindo a paleta de cores
ax = sns.countplot(x = x, data = y, order = ordenando, palette = 'cool_r')
#Adicionando título ao gráfico e definindo tamanho
ax.set_title('Top 5 compostos', {'fontsize': 26})
#Definindo título e tamanho dos eixos
plt.xlabel("Composto", size=15)
plt.ylabel("Quantidade", size= 15)
plt.show()
```

***Desafio 03: Plotar os histogramas com seaborn.***

```python
#Definindo G
g = dados.loc[:, 'g-0':'g-771'].describe().T
```

```python
#Faznedo a média do std de G
g['std'].mean()
```

```python
#Definindo C
c = dados.loc[:, 'c-0': 'c-99'].describe().T
```

```python
#Faznedo a média do std de C
c['std'].mean()
```

```python
#Utilizamos essa função para salvar o padrão de imagens do seaborn
sns.set()
#Definindo tamanho do gráfico
plt.figure(figsize = (14,6))
#Utilizamos essa função para podermos plotar mais de um gráfico
plt.subplot(1,2,1)
#A função kde adiciona uma linha de tendência
sns.histplot(x = 'std', kde = True, data = g, bins = 30, color = 'indigo')
#Utilizamos essa função para podermos plotar mais de um gráfico
plt.subplot(1,2,2)
#A função kde adiciona uma linha de tendência
sns.histplot(x = 'std', data = c, kde = True, bins = 30, color = 'springgreen')
plt.show()
```

***Desafio 06: Fazer outras análises com o boxplot e até com o histograma.***

```python
#Utilizamos a função subplots para conseguirmos visualizar dois gráficos lado a lado
fig, axes = plt.subplots(1, 2, figsize=(18, 8), sharey=True)
#Definindo paramêtros do gráfico 1 (dose)
#A função axes permite definirmos os eixos
sns.boxplot(ax=axes[0], y = 'g-7', x ='dose' , data = dados, palette = 'flare')
#Definindo título e tamanho da fonte
axes[0].set_title('Análise Dose', fontsize=15)
#Definindo título dos eixos e tamanho da fonte
axes[0].set_xlabel('Doses',fontsize=12)
axes[0].set_ylabel('Valores G-7',fontsize=12)
#Definindo paramêtros do gráfico 2 (tempo)
sns.boxplot(ax=axes[1], y = 'g-7', x = 'tempo' , data = dados, palette = 'bone')
axes[1].set_title('Análise Tempo', fontsize=15)
axes[1].set_xlabel('Tempo',fontsize=12)
axes[1].set_ylabel('Valores G-7',fontsize=12)
```

```python
#Utilizamos a função subplots para conseguirmos visualizar dois gráficos ao mesmo tempo
#A função ax permite definirmos os eixos
fig, ax = plt.subplots(2, constrained_layout=True, figsize =(13,8))
#Definindo paramêtros do gráfico 1 (g)
ax1 = sns.histplot(data = dados.loc[:,'g-0':'g-771'].describe().T['mean'], kde = True , bins = 50, ax=ax[0], color = 'deeppink')
ax[0].set_xlabel('Média', fontsize=15)
ax[0].set_ylabel('Quantidade', fontsize=15)
ax[0].set_title('Histograma G', family='san-serif', fontsize=20)
#Definindo paramêtros do gráfico 2 (c)
ax2 = sns.histplot(data = dados.loc[:,'c-0':'c-99'].describe().T['mean'], kde = True , bins = 50, ax=ax[1], color = 'turquoise')
ax[1].set_xlabel('Média', fontsize=15)
ax[1].set_ylabel('Quantidade', fontsize=15)
ax[1].set_title('Histograma C', family='san-serif', fontsize=20)
plt.show()
```
